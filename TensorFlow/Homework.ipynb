### Newton's method

Find roots with **Newton's method** for:


1) $\:f(x) = x\,cos(x) - x^2 \:\:\:  x_0 = 1$

2) $\:f(x) = x^3 - 7x^2 + 8x - 3 \:\:\: x_0 = 5 $

```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
config.gpu_options.per_process_gpu_memory_fraction = 0.1
```

```python
sess = tf.Session(config=config)
X = np.linspace(-2, 6, 100)
```

```python
x = tf.placeholder(tf.float32, name="x")
y_1 = x*tf.cos(x) - x**2
y_2 = x**3 - 7*x**2 + 8*x - 3
```

```python
grad_y_1 = tf.gradients(y_1, x, name="grad_y_1")
grad_y_2 = tf.gradients(y_2, x, name="grad_y_2")
```

```python
plt.figure(figsize=(12, 8))

plt.plot(X, sess.run(y_1, {x: X}), label="function1",c='green')
plt.plot(X, sess.run(y_2, {x: X}), label="function2",c='blue')
plt.axhline(y=0, linewidth=1)
plt.legend();
```

```python
iterations = 10
x_1 = [1]

for _ in range(iterations):
    current_x = x_1[-1]
    next_x = current_x - sess.run(y_1/grad_y_1[0], {x: current_x})
    x_1.append(next_x)
    
x_2 = [5]

for _ in range(iterations):
    current_x = x_2[-1]
    next_x = current_x - sess.run(y_2/grad_y_2[0], {x: current_x})
    x_2.append(next_x)
```

```python
plt.figure(figsize=(12, 8))

plt.plot(X, sess.run(y_1, {x: X}), label="function1",c='green')
# plt.plot(X, sess.run(y_2, {x: X}), label="function2",c='blue')
plt.scatter(x_1, sess.run(y_1, {x: x_1}), s = 50/np.arange(1, 1+len(x_1)), c="red")


# plt.scatter(x_2, sess.run(y_2, {x: x_2}), s = 50/np.arange(1, 1+len(x_2)), c="yellow")

plt.axhline(y=0, linewidth=1)
```

```python
plt.figure(figsize=(12, 8))

# plt.plot(X, sess.run(y_1, {x: X}), label="function1",c='green')
plt.plot(X, sess.run(y_2, {x: X}), label="function2",c='blue')
# plt.scatter(x_1, sess.run(y_1, {x: x_1}), s = 50/np.arange(1, 1+len(x_1)), c="red")


plt.scatter(x_2, sess.run(y_2, {x: x_2}), s = 50/np.arange(1, 1+len(x_2)), c="black")

plt.axhline(y=0, linewidth=1)
```

### Softmax

$$\text{softmax}(x)_i = \frac{\exp(x_i)}{\sum_j \exp(x_j)}$$

```python
def softmax(sess, x):
    """
    param sess: tf.Session
    param x:  np.array, n-dimentional vector
    returns: np.array n-dimentional vector
    """
    def soft(x):
        sm = np.sum(np.exp(x))
        return np.exp(x)/sm
    
    data = tf.placeholder(tf.float32, name='data')
    ret = tf.py_func(func=soft,inp=[x],Tout=tf.double,name='ret')
    return sess.run(ret, {data:x})
```

```python
# test
import tensorflow as tf
import numpy as np
```

```python
with tf.Session() as sess:
    if np.allclose([0.0320586, 0.08714432, 0.23688282, 0.64391426],
                   softmax(sess, np.array([1, 2, 3, 4]))):
        print("Test passed!")
    else:
        print("Test failed!")
```

$$\text{softmax}(X)_{ij} = \frac{\exp(X_{ij})}{\sum_j \exp(X_{ij})}$$

```python
def softmax_matrix(sess, x):
    """
    Computes softmax for each row vector in matrix.
    
    param sess: tf.Session
    param x:  np.array, (n x m) matrix
    returns: np.array, (n x m) matrix
    """
    def soft_matrix(x):
        softs = []
        for d in x:
            sm = np.sum(np.exp(d))
            softs.append(np.exp(d)/sm)
        return np.array(softs)
    
    data = tf.placeholder(tf.float32,name='data')
    ret = tf.py_func(func=soft_matrix,inp=[x],Tout=tf.double,name='ret')
    output = sess.run(ret, {data:x})
    return output
```

```python
with tf.Session() as sess:
    if np.allclose([[0.10650698, 0.78698604, 0.10650698],
                    [0.09003057, 0.24472847, 0.66524096]],
                   softmax_matrix(sess, np.array([[1,3,1], [-1, 0, 1]]))):
        print("Test passed!")
    else:
        print("Test failed!")
```
